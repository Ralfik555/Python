```python
# Let's create two distributions
data1 = randn(100)
data2 = randn(100) + 2 # Off set the mean
```

```python
sns.boxplot(data = [data1,data2])
```

```python
sns.boxplot(data = [data1,data2],whis = np.inf)
```

```python
sns.boxplot(data = [data1,data2],orient='h')
```

```python
data1 = stats.norm(0,5).rvs(100)

# Two gamma distributions concatenated together (Second one is inverted)
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                        -1*stats.gamma(5).rvs(50)])
```

```python
sns.boxplot(data=[data1,data2],whis=np.inf)
```

```python
sns.violinplot(data=[data1,data2],whis=np.inf)
```

```python
sns.violinplot(data2,bw=.02)
```

```python
sns.violinplot(data1,inner='stick')
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
tips[tips['smoker']=='Yes'].count()
```

```python
tips['day'].unique()
```

```python
sns.lmplot("total_bill","tip",tips)
```

```python
sns.lmplot("total_bill","tip",tips,scatter_kws = {'marker':'o','color':'indianred'},
           line_kws={'linewidth':1,'color':'blue'})
```

```python
sns.lmplot("total_bill","tip",tips,order = 3,
           scatter_kws = {'marker':'o','color':'indianred'},
           line_kws={'linewidth':1,'color':'blue'})
```

```python
sns.lmplot("total_bill","tip",tips,fit_reg=False)
```

```python
tips['tip_perc'] = round(100*tips.tip/tips.total_bill,2)
```

```python
tips.head()
```

```python
sns.lmplot('size','tip_perc',tips)
```

```python
sns.lmplot('size','tip_perc',tips,x_jitter=.1)
```

```python
sns.lmplot('size','tip_perc',tips,x_estimator=np.mean)
```

```python
sns.lmplot("total_bill", "tip_perc", tips,hue='sex',markers=['x','o'])
```

```python
sns.lmplot("total_bill", "tip_perc", tips,hue='day')
```

```python
sns.lmplot("total_bill", "tip_perc", tips,lowess=True,line_kws={'color':'black'})
```

```python
sns.regplot("total_bill", "tip_perc", tips)
```

```python
fig, (axis1,axis2) = plt.subplots(2,1,sharey =True)

sns.regplot("total_bill","tip_perc",tips,ax=axis1)
sns.violinplot(tips['size'],tips['tip_perc'],color='indianred',ax=axis2)
```

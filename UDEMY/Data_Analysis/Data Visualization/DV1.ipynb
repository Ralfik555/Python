```python
#Kernel Density Estimation plot
```

```python
dataset = randn(25)
```

```python
sns.rugplot(dataset)
plt.ylim(0,1)
```

```python
plt.hist(dataset,alpha = 0.3)
sns.rugplot(dataset)
```

```python
sns.rugplot(dataset)
x_min,x_max = dataset.min() - 2, dataset.max() + 2
x_axis = np.linspace(x_min,x_max,num=100)
band_width = ((4*(dataset.std())**5)/(3*len(dataset)))**0.2
kernel_list = []

for datapoint in dataset:
    kernel = stats.norm(datapoint,band_width).pdf(x_axis)
    kernel_list.append(kernel)
    
    kernel = kernel / kernel.max()
    kernel = kernel * 0.4
    
    plt.plot(x_axis,kernel,alpha = 0.5,color='grey')
    

plt.ylim(0,1)
sum_kernel_list = sum(kernel_list)    
```

```python
fig = plt.plot(x_axis,sum_kernel_list,color= 'indianred')
sns.rugplot(dataset)
plt.yticks([])
plt.suptitle('Sum of basic functions')
```

```python
# Mean center of data
mean = [0,0]

# Diagonal covariance
cov = [[1,0],[0,100]]

# Create dataset using numpy
dataset2 = np.random.multivariate_normal(mean,cov,1000)

# Bring back our old friend pandas
dframe = pd.DataFrame(dataset2,columns=['X','Y'])
```
